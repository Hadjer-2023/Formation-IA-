### Part 1: using beautiful soupe

# Import modules: 

```python
from bs4 import BeautifulSoup
import requests
import re
```

# Get the data using url: 

```python
url = "https://www.linkedin.com/feed/?highlightedUpdateUrn=urn%3Ali%3Aactivity%3A7190666498305142784"
```

```python
# get the page data:

page = requests.get(url)
```

```python
page.content
```

```python
# parse the data:

soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

# Regex de find particular class:

```python
tag = soup.find('div', class_=re.compile(r'jo+'))
print(tag)
```

```python
tag.text
```
